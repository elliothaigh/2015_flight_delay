### Merging databases

```python
# combine df_flights and df_airlines
merged_df = pd.merge(df_flights, df_airlines, left_on='AIRLINE', right_on = 'IATA_CODE')
```

```python
merged_df.head()
```

```python
# clean column names
merged_df = merged_df.drop(columns = ['AIRLINE_x', 'FLIGHT_NUMBER', 'TAIL_NUMBER', 'IATA_CODE'])
```

```python
# combine df_airports with merged_df (origin airport)
merged_df2 = pd.merge(merged_df, df_airports, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', how='outer')
```

```python
merged_df2.head()
```

```python
# combine df_airports with merged_df (destination airport)
merged_df2 = pd.merge(merged_df2, df_airports, left_on='DESTINATION_AIRPORT', right_on='IATA_CODE', how='outer')
```

```python
merged_df2.head()
```

### Drop columns

```python
# Check columns
for col in merged_df2:
    print(col)
```

```python
# We do not need some columns for our analysis and therefore will remove them from the dataframe
merged_df2 = merged_df2.drop(columns = ['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'])
```

### Renming Columns

```python
dict = {'AIRLINE_y' : 'AIRLINE',
        'IATA_CODE_x':'ORIGIN_AIRPORT_CODE',
        'AIRPORT_x':'ORIGIN_AIRPORT',
        'CITY_x':'ORIGIN_CITY',
        'STATE_x' : 'ORIGIN_STATE',
        'COUNTRY_x': 'ORIGIN_COUNTRY',
        'LATITUDE_x': 'ORIGIN_LATITUDE',
        'LONGITUDE_x': 'ORIGIN_LONGITUDE',
       'IATA_CODE_y':'DESTINATION_AIRPORT_CODE',
        'AIRPORT_y':'DESTINATION_AIRPORT',
        'CITY_y':'DESTINATION_CITY',
        'STATE_y' : 'DESTINATION_STATE',
        'COUNTRY_y': 'DESTINATION_COUNTRY',
        'LATITUDE_y': 'DESTINATION_LATITUDE',
        'LONGITUDE_y': 'DESTINATION_LONGITUDE'}
```

```python
merged_df2.rename(columns=dict,
          inplace=True)
```
